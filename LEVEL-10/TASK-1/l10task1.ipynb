# Descriptive statistics 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
data= pd.read_csv("Titanic-Dataset.csv")
```

```python
data.head()
```

```python
Mean_age= data["Age"].mean()
median_age = data['Age'].median()
mode_age = data['Age'].mode()[0]
```

```python
Mean_Fare= data["Fare"].mean()
median_Fare = data['Fare'].median()
mode_Fare = data['Fare'].mode()[0]
```

```python
print(f"Mean Age: {Mean_age}, Median Age: {median_age}, Mode Age: {mode_age}")
print(f"Mean Fare: {Mean_Fare}, Median Fare: {median_Fare}, Mode Fare: {mode_Fare}")
```

```python
survival_by_gender= data.groupby("Sex")["Survived"].mean()*100
print("survival_by_gender:",survival_by_gender)
```

```python
survival_by_Pclass= data.groupby("Pclass")["Survived"].mean()*100
print("survival_by_Pclass:",survival_by_Pclass)
```

```python
survival_by_embarked=data.groupby("Embarked")["Survived"].mean()*100
print("survival_by_embarked:",survival_by_embarked)
```

```python
corr_matrix = data.corr(numeric_only=True)
corr_matrix
```

```python
plt.figure(figsize=(8, 5))
colors = ["#ff9999", "#66b3ff"]
sns.barplot(x=survival_by_gender.index, y=survival_by_gender.values, palette=colors)
plt.title("Survival Rate by Gender", fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Survival Rate (%)", fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
colors = ["#ffcc99", "#99ff99", "#6699ff"]
sns.barplot(x=survival_by_Pclass.index, y=survival_by_Pclass.values, palette=colors)
plt.title("Survival Rate by Passenger Class", fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Survival Rate (%)", fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
colors = ["#ff6666", "#ffcc66", "#66ccff"]
sns.barplot(x=survival_by_embarked.index, y=survival_by_embarked.values, palette=colors)
plt.title("Survival Rate by Embarkation Port", fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel("Embarkation Port", fontsize=12)
plt.ylabel("Survival Rate (%)", fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap="Spectral", fmt=".2f", linewidths=0.5, linecolor='black', cbar_kws={'shrink': 0.75})
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold', color='darkblue')
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
