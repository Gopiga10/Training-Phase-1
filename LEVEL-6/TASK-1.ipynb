## Descriptive Statistics 

Describe how it summarizes and organizes data using measures like mean, median, mode, and graphs.

```python
import numpy as np
import scipy.stats as stats
```

```python
num=np.random.randint(20,40,10)
```

```python
num
```

```python
print("Mean:",np.mean(num))
```

```python
print("Median:",np.median(num))
```

```python
print("Standard_deviation:",np.std(num))
```

```python
print("variance:",np.var(num))
```
