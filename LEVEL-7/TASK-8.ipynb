```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 10  
p = 0.5 
size = 1000  
```

```python
binomial_samples = np.random.binomial(n=n, p=p, size=size)
binomial_samples
```

```python
simulated_mean = np.mean(binomial_samples)
simulated_mean
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(binomial_samples,  bins='auto', kde=False, color="red", edgecolor="black") 
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title(f"Binomial Distribution (n={n}, p={p})\nSimulated Mean: {simulated_mean:.2f}")
plt.xticks(range(n+1))
plt.show()
```

```python
print(f"Simulated Mean of Distribution: {simulated_mean:.3f}")
```
