```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass

```

```python
df = pd.read_csv("House Price Prediction Dataset.csv")
```

```python

@dataclass
class DatasetStats:
    mean: float
    max: float
    min: float

    def plot_histogram(self, df, column_name):
        plt.figure(figsize=(12, 6)) 
        sns.set_style("whitegrid")
        sns.histplot(df[column_name], bins=30, kde=True, color="royalblue", alpha=0.75, edgecolor="black")
        mean_value = df[column_name].mean()
        median_value = df[column_name].median()
        plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f' Mean: {mean_value:,.2f}')
        plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label=f' Median: {median_value:,.2f}')
        plt.xlabel(f"{column_name} ($)", fontsize=14, fontweight="bold", color="darkblue")
        plt.ylabel("Frequency", fontsize=14, fontweight="bold", color="darkblue")
        plt.title(f" Distribution of {column_name}", fontsize=18, fontweight="bold", color="darkred")
        plt.legend(fontsize=12)
        plt.show()
```

```python
print("Available Columns:", list(df.columns))
column_name = input("Enter the column name to visualize: ")
if column_name in df.columns:
    stats = DatasetStats(
        mean=df[column_name].mean(),
        max=df[column_name].max(),
        min=df[column_name].min(),
    )
    stats.plot_histogram(df, column_name)
else:
    print(" Invalid column name. Please run the script again with a valid column.")
```
